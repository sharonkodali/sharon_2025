- **Take note and describe the type of shell commands you are using through Terminal in this installation procedure.** This is specific to your machine.

 # 'cd'
  - Usage: cd path/to/directory
  - Description: Changes the current working directory to the specified path. For example, cd Documents moves into the "Documents" directory. This command is essential for navigating the file system.

 2. ls (List Directory Contents)
  - Usage: ls, ls -l, ls -a
  - Description: Lists files and directories in the current directory. Options like -l provide a detailed listing (including file permissions and sizes), and -a includes hidden files.
3. pwd (Print Working Directory)
  - Usage: pwd
  - Description: Displays the full path of the current working directory. This command helps you know where you are in the file system.
4. mkdir (Make Directory)
  - Usage: mkdir new_directory
  - Description: Creates a new directory with the specified name. For example, mkdir project creates a new directory named "project."
5. rm (Remove)
- Usage: rm file, rm -r directory
- Description: Deletes files or directories. The -r option allows recursive deletion of directories and their contents.
6. brew (Homebrew)
- Usage: brew install package, brew update
- Description: A package manager for macOS that simplifies the installation and management of software. For example, brew install python installs Python, and brew update updates Homebrew and its packages.
7. pip (Python Package Installer)
- Usage: pip install package_name
- Description: Installs Python packages from the Python Package Index (PyPI). For example, pip install numpy installs the NumPy package.
8. python (Python Interpreter)
- Usage: python script.py, python -m module
- Description: Executes Python scripts or runs Python modules. For instance, python my_script.py runs a Python script named my_script.py.
9. make
- Usage: make target
- Description: Runs commands defined in a Makefile, typically used for building and managing projects. For example, make build executes the commands to build a project as specified in the Makefile.
10. source
- Usage: source file.sh
- Description: Executes commands from a file in the current shell session. Often used to set environment variables or execute scripts.
11. export
- Usage: export VAR=value
- Description: Sets environment variables that can be used by other commands and scripts. For example, export PATH=$PATH:/new/path adds a new directory to the system PATH

In the development process, tools like GitHub and GitHub Pages are essential for version control and deploying websites. Locally, you run and view your project through a localhost URL (e.g., http://localhost:4000), which is only visible to you. This is useful for testing and development. In contrast, when your project is deployed on GitHub Pages, it becomes accessible via a public URL (e.g., https://username.github.io/repository/), allowing anyone to view your site online.

DNS (Domain Name System) plays a crucial role in web addresses. By default, GitHub Pages provides a URL under github.io. If you set up a custom domain, DNS helps direct traffic from your chosen domain name (e.g., https://www.yourdomain.com) to your GitHub Pages site. This allows for a more personalized and professional web address.

In summary, localhost is for private development, GitHub Pages is for public deployment, and DNS manages how users access your site via web addresses.
